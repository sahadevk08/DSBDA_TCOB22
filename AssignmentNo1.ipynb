```python
from sklearn import preprocessing
df = pd.read_csv(r"C:\Users\sahad\Downloads\iris.csv")
df.head()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x=df.iloc[:,:4]
x_scaled = min_max_scaler.fit_transform(x)
df_normalized = pd.DataFrame(x_scaled)
df_normalized
```

```python
df['species'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
df['species']= label_encoder.fit_transform(df['species'])
df['species'].unique()
```

```python
from sklearn import preprocessing

# Initialize the OneHotEncoder
enc = preprocessing.OneHotEncoder()

# Fit and transform, then convert the result to a dense array and into a DataFrame
enc_df = pd.DataFrame(enc.fit_transform(df[['species']]).toarray())

# View the encoded DataFrame
enc_df
```

```python
# Assuming 'df' is your original DataFrame and 'species' is the target column
features_df = df.drop(columns=['species'])

# Then join the encoded 'species' DataFrame (enc_df) to 'features_df'
df_encode = features_df.join(enc_df)

# Optionally, rename the encoded columns to match the species names
df_encode.rename(columns={0:'Iris-Setosa', 1:'Iris-Versicolor', 2:'Iris-virginica'}, inplace=True)

# View the encoded DataFrame
df_encode
```
